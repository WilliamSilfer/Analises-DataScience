```python
import pandas as pd
```

```python
# Criando uma tabela de frequencias
# 1 para Estatístico
# 2 para Cientista de Dados
# 3 para Programador Python
```

```python
dados = pd.DataFrame({'Profissão': [1, 2, 3, 1, 2, 2, 2, 3, 3, 2, 1, 3]})
```

```python
dados['Profissão'].value_counts()
```

```python
percentual = dados['Profissão'].value_counts(normalize = True) * 100
```

```python
percentual
```

```python
frequencia = dados['Profissão'].value_counts()
```

```python
df = pd.DataFrame({'Frequencia': frequencia, 'Porcentual (%)': percentual})
```

```python
df
```

```python
df.rename(index = {2: 'Cientista de Dados', 3: 'Programador Python', 1: 'Estatístico'})
```

```python
df.rename_axis('Profissão', axis = 'columns', inplace = True)
```

```python
df
```

```python
import seaborn as sns
import matplotlib as plt
```

```python
# Pré visualização dos dados
sns.barplot(x = 'Frequencia', y = 'Porcentual (%)', data = df)
```
